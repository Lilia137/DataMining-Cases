```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_csv("goods.csv")
data
```

```python
freq_set=apriori(data,min_support=0.5,use_colnames=True)
freq_set['num']=freq_set['support']*4
freq_set
```

```python
rules = association_rules(freq_set,metric='confidence',min_threshold=0.7)
rules
```
