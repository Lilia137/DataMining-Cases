```python
# 二分聚类算法
# 为了克服k-均值算法的局部最小值问题
def compute_sub_sse(data, k, kmeans):
    cluster_sse = np.zeros(k)
    for i in range(0,len(data)):
        label = data.iloc[i,len(data.columns)-1]
#         print(label)
        point = data.iloc[i,:len(data.columns)-1]
        cluster_center = kmeans.cluster_centers_[label]
#         print(len(point))
#         print(len(cluster_center))
        cluster_sse[label] += np.square(point - cluster_center).sum()
    return cluster_sse
```

```python
# 可视化需要：每组数据的label，求整体的SSE需要每组的中心点，
def biKmeans(data, k):
    cluster_num = 0      # 簇的个数
    result = pd.DataFrame(columns = new_df.columns)   # 最终结束后的带标签数据集
    total_sse = 0    # 最终结束后的sse
    while cluster_num+2 <= k:

        km_f = KMeans(n_clusters = 2, random_state = 10).fit(data)
        
#         将label加进去
        data["jllable"] = km_f.labels_ 
        
#         计算父节点的SSE，（1，2）数组
        cluster_sse = compute_sub_sse(data, 2, km_f)
        data["jllable"] = km_f.labels_ + cluster_num
        print('第'+str(cluster_num)+'次划分\n',km_f.labels_ + cluster_num)    
        
#         如果到了终止条件
        if cluster_num+2 == k:
            total_sse += km_f.inertia_
            result = result.append(data)
            cluster_num += 1
        else:
    #         对子节点进行二分聚类
            data_0 = data[data["jllable"] == cluster_num]
            data_1 = data[data["jllable"] == cluster_num+1]
            if len(data_0)>=2:
                km_c0 = KMeans(n_clusters = 2, random_state = 10).fit(data_0.drop('jllable',axis = 1))
                dssm_0 = cluster_sse[0] - km_c0.inertia_
            else:
                dssm_0 = 0
            if len(data_1)>=2:
                km_c1 = KMeans(n_clusters = 2, random_state = 10).fit(data_1.drop('jllable',axis = 1))
                dssm_1 = cluster_sse[1] - km_c1.inertia_
            else:
                dssm_1 = 0
            
    #         比较两个簇的SSE差值，选差值大的那个进行划分
            if dssm_0>=dssm_1:
                total_sse += cluster_sse[1]
                data_1["jllable"] = cluster_num
                result = result.append(data_1)
                data = data_0.drop('jllable',axis = 1)
                cluster_num += 1
            else:
                total_sse += cluster_sse[0]
                result = result.append(data_0)
                data = data_1.drop('jllable',axis = 1)
                cluster_num += 1
    return total_sse, result 
```

```python
# 初始化二分k均值所用的数据集
bi_data = data.copy()
# total_sse, result = biKmeans(bi_data, 5)
```

```python
SSE = []
SSE_bi = []
for i in range(2,8):
    bi_data = data.copy()
    km = KMeans(n_clusters = i, random_state = 10).fit(data)
    SSE.append(km.inertia_)
    
    total_sse, result = biKmeans(bi_data, i)
    SSE_bi.append(total_sse)
myfont = fm.FontProperties('simhei',size=12.5)
plt.plot(range(2,8),SSE,marker="o",label = 'binary_kmeans')
plt.plot(range(2,8),SSE_bi,marker="*", label = 'kmeans')
plt.xlabel("K值—簇数量",fontproperties=myfont)
plt.ylabel("SSE",fontproperties=myfont)
plt.legend(loc = 'best')
plt.show()
```
